# MERGE ALLRECIPES DATA WITH CATEGORIES AND CALORIES


```python
%matplotlib inline 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import bokeh
from bokeh.io import output_notebook, show
output_notebook()

import pandas as pd
print ("Pandas version: {}".format(pd.__version__))
```

```python
recipes_orig = pd.read_csv('allrecipes.csv', sep=",", error_bad_lines=False, encoding="latin-1")
```

```python
'num of rating and num of fields' 
recipes_orig['url'] = recipes_orig['url'].str.lower()
print(recipes_orig.shape)
print(list(recipes_orig.columns))
recipes_orig.head()
```

```python
extra_data = pd.read_csv('extra_data.csv', sep=",", error_bad_lines=False, encoding="latin-1")
```

```python
'num of rating and num of fields' 
print(extra_data.shape)
print(list(extra_data.columns))
extra_data.head()
```

```python
cal_cat = extra_data.filter(['calories', 'category', 'url'], axis=1)
```

```python
print(cal_cat.shape)
print(list(cal_cat.columns))
cal_cat.head()
```

```python
result_data = pd.merge(recipes_orig,
                      cal_cat[['calories', 'category', 'url']],
                      on='url')
```

```python
print(result_data.shape)
print(list(result_data.columns))
result_data.head()
```
